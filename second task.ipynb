```python
import numpy as np
import scipy as sp
import math as m
```

```python
def func(x):    
    return m.sin(x/5.) * m.exp(x/10.) + 5. * m.exp(-x/2.)
```

```python
x_arr = range(1,16)
y_arr = [func(float(x)) for x in x_arr]
```

```python
%matplotlib inline
plot(x_arr,y_arr)
```

```python
from scipy import linalg

a = np.array([[1, 1], [1, 15]])
b = []
b.append(func(float(1)))
b.append(func(float(15)))
x = linalg.solve(a,b)

print x
```

```python
y_arr2 = [x[0] + x[1]*item for item in x_arr]
plot(x_arr,y_arr2)
plot(x_arr,y_arr)
    
```

```python
a = np.array([[1, 1, 1], [1, 8, 64], [1, 15, 15*15]])
b = []
b.append(func(1))
b.append(func(8))
b.append(func(15))
x2 = linalg.solve(a,b)

print x2
```

```python
y_arr3 = [x2[0] + x2[1]*item + x2[2]*item**2 for item in x_arr]
plot(x_arr,y_arr2)
plot(x_arr,y_arr)
plot(x_arr,y_arr3)
```

```python
a = np.array([[1, 1, 1, 1], [1, 4, 16, 64], [1, 10, 100, 1000], [1, 15, 15*15, 15*15*15]])
b = []
b.append(func(1))
b.append(func(4))
b.append(func(10))
b.append(func(15))
x3 = linalg.solve(a,b)
a
```
